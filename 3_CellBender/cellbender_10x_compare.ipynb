```python
import scanpy as sc
import warnings
import pandas as pd
from os import listdir
from os.path import isfile, join
from matplotlib import pyplot as plt
from pandas.plotting import table
```

```python
# https://stackoverflow.com/questions/3207219/how-do-i-list-all-files-of-a-directory
mypath = "output_files/"
onlyfiles = [f for f in listdir(mypath)]
print(onlyfiles)
```

```python
warnings.filterwarnings('ignore')

cellbender_n_genes = []
cellbender_n_cells = []

tenx_n_genes = []
tenx_n_cells = []

# https://scanpy.readthedocs.io/en/stable/tutorials/basics/clustering.html
for file in onlyfiles:

    cellbender_name = f"{mypath}{file}/cellbender_output_file_filtered.h5"
    cellbender = sc.read_10x_h5(cellbender_name)
    cellbender.var_names_make_unique()
    
    tenx_name = f"{mypath}{file}/filtered_feature_bc_matrix.h5"
    tenx = sc.read_10x_h5(tenx_name)
    tenx.var_names_make_unique()
    
    #"""
    # mitochondrial genes, "MT-" for human, "Mt-" for mouse
    cellbender.var["mt"] = cellbender.var_names.str.startswith("MT-")
    # ribosomal genes
    cellbender.var["ribo"] = cellbender.var_names.str.startswith(("RPS", "RPL"))
    # hemoglobin genes
    cellbender.var["hb"] = cellbender.var_names.str.contains("^HB[^(P)]")
    
    sc.pp.calculate_qc_metrics(
        cellbender, qc_vars=["mt", "ribo", "hb"], inplace=True, log1p=True
    )
    
    # mitochondrial genes, "MT-" for human, "Mt-" for mouse
    tenx.var["mt"] = tenx.var_names.str.startswith("MT-")
    # ribosomal genes
    tenx.var["ribo"] = tenx.var_names.str.startswith(("RPS", "RPL"))
    # hemoglobin genes
    tenx.var["hb"] = tenx.var_names.str.contains("^HB[^(P)]")
    
    sc.pp.calculate_qc_metrics(
        tenx, qc_vars=["mt", "ribo", "hb"], inplace=True, log1p=True
    )
    #"""

    sc.pp.filter_cells(cellbender, min_genes=100)
    sc.pp.filter_genes(cellbender, min_cells=3)
    
    sc.pp.filter_cells(tenx, min_genes=100)
    sc.pp.filter_genes(tenx, min_cells=3)
    
    # Saving count data
    cellbender.layers["counts"] = cellbender.X.copy()
    # Normalizing to median total counts
    sc.pp.normalize_total(cellbender)
    # Logarithmize the data
    sc.pp.log1p(cellbender)
    
    # Saving count data
    tenx.layers["counts"] = tenx.X.copy()
    # Normalizing to median total counts
    sc.pp.normalize_total(tenx)
    # Logarithmize the data
    sc.pp.log1p(tenx)
    
    cellbender_n_genes.append(len(cellbender.obs))
    cellbender_n_cells.append(len(cellbender.var))

    tenx_n_genes.append(len(tenx.obs))
    tenx_n_cells.append(len(tenx.var))
    
    
    print(file)
    print("cellbender:")
    print(cellbender)
    sc.pl.violin(cellbender, ["Col1a1", "Postn", "Gja1"], stripplot=False, ylabel = "normalized gene expression", save=f"{file}_cellbender.pdf")
    print("tenx:")
    print(tenx)
    sc.pl.violin(tenx, ["Col1a1", "Postn", "Gja1"], stripplot=False, ylabel= "normalized gene expression", save=f"{file}_tenx.pdf")
    print("")
    print("")
    
    
```

```python
pd.set_option('display.width', 200)
d = {'sample': onlyfiles, '10x_n_cells': tenx_n_cells, 'cellbender_n_cells': cellbender_n_cells, 
     '% less cells': ([100*((a_i - b_i)/a_i) for a_i, b_i in zip(tenx_n_cells, cellbender_n_cells)]),
     '10x_n_genes': tenx_n_genes, 'cellbender_n_genes': cellbender_n_genes, 
     '% less genes': ([100*((a_i - b_i)/a_i) for a_i, b_i in zip(tenx_n_genes, cellbender_n_genes)])
     }
df = pd.DataFrame(data = d)
print(df)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

genes = ["Col1a1", "Postn", "Gja1"]

# Extract expression data for each gene
df_cellbender = cellbender[:, genes].to_df()
df_cellbender["source"] = "cellbender"

df_tenx = tenx[:, genes].to_df()
df_tenx["source"] = "10x"

# Combine them
df_combined = pd.concat([df_cellbender, df_tenx])

# Melt to long format for seaborn
df_melted = df_combined.melt(id_vars="source", 
                             var_name="gene", 
                             value_name="expression")


```

```python
plt.figure(figsize=(8, 5))
sns.violinplot(
    data=df_melted,
    x="gene", 
    width = 0.8,
    y="expression", 
    hue="source",
    scale="count",
    split=True,      # side-by-side, not overlapped
    inner=None,      # adds a box inside for median/IQR
    cut=0,            # avoids extending beyond data range
)
sns.despine()
plt.ylabel("Normalized Gene Expression")
plt.xlabel("")
plt.title("Cellbender vs 10x Marker Gene Expression")
plt.legend(title="")
plt.tight_layout()
plt.savefig(f"{file}_violin_comparison.pdf")
plt.show()
```
